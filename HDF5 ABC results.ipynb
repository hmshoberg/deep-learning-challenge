```python
import h5py

def print_hdf5_items(item, item_name=""):
    """
    Recursively print information about datasets within groups in an HDF5 file.
    """
    if isinstance(item, h5py.Dataset):
        # If it's a dataset, print its name, shape, and a snippet of the content
        print(f"Dataset: {item_name}, Shape: {item.shape}")
        print(item[:])  # Displaying the content (adjust as needed)
    elif isinstance(item, h5py.Group):
        # If it's a group, iterate through its items and call the function recursively
        print(f"Group: {item_name}")
        for sub_item_name, sub_item in item.items():
            full_name = f"{item_name}/{sub_item_name}" if item_name else sub_item_name
            print_hdf5_items(sub_item, full_name)

# Replace 'your_file.h5' with the actual path to your HDF5 file
file_path = r'X:\SCHOOL\UofM BootCamp\Module-21-Challenge\deep-learning-challenge\AlphabetSoupCharity.h5'

# Open the HDF5 file in read-only mode
with h5py.File(file_path, 'r') as file:
    # Call the recursive function starting from the root of the file
    print_hdf5_items(file)
```
